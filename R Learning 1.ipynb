* append: It defines that the exported data is going to be stored into a new file or in existing file. If it is true then our data will be written in an existing file from the last line of the file. it it is false then it will rewrite an existing from the beggining or it will create a new file.

### 1. tapply function.

This function is used to apply a function on a data with a particular type of group. Suppose that we have a time series data and we want to calculate the mean on the data by per month then in this case we need to use tapply which will apply mean fucntion on all the month groups.

**str**: This function is used to compactly display the internal structure of an R object, a diagnostic function and an alternative to summary.

```r
# For this  example we will import a datase of vegetation2 data set 
veg = read.table(file="RBook/RBook/vegetation2.txt", header=TRUE)
```

```r
# after loading the data we need to check head and str on the data
head(veg)
```

```r
# Now we will use the str() method on the imported dataset
str(veg)
```

As we can say that in this dataset we have three type of data, Factor, int and num. To avoid the Factor type of data during the importation please use the right seperator in read.table() method.

#### Task1: Now we need to calculate Mean of R per Transect

**=>** Now in this case we need to calculate the mean of R, but the situation is that we need to calculate the mean of R per Transect, for this we will use the tapply function.

```r
# Calculating mean of R per Transect
Rmeans = tapply(veg$R, veg$Transect, mean)
```

```r
Rmeans
```

In above output we can see the mean of the R data by per Transect, In this way we can perform any kind of function on the data.

#### Task2: Make a dataset of mean, median, std, length of R data by per Transect

```r
# again we will use the tapply function to do this but in this case we will apply it 4 time for each fucntion.
Rmedian = tapply(veg$R, veg$Transect, median)
Rstd = tapply(veg$R, veg$Transect, sd)
Rlength = tapply(veg$R, veg$Transect, length)
```

```r
# Now we will bind them by using the cbind function 
Rsummary = cbind(Rmeans, Rmedian, Rstd, Rlength)
Rsummary
```

This table can be usefull to see the spread of the data over per Transect and deviation from the mean.

## 2. sapply and lapply function.

`sapply:` This function is used to apply a function over a list or vector not on the group of the data. by default it returns a vector, matrix or if simplify = "array", an array if appropriate, by applying simplify2array().


`lapply:` This function is similar to the sapply function this returns a list.

#### Task1: Calculate the mean of each data having int type.

Suppose that we want to canculate the mean of all the data available in a coulmn and we want to do this for many column then we can use sapply and lapply function.

```r
# to apply a function on a dataset first of all we need to combine them by cbind or matrix or a dataframe.
# in this case we will use cbind on which we want to calculate the mean of the data.
Datasets = cbind(veg$R, veg$ROCK, veg$LITTER, veg$ML, veg$BARESOIL)
```

```r
Datasets[1:10,]
```

```r
# Now we will apply the function on this data.
veg_means = sapply(Datasets, FUN=mean)
```

```r
veg_means
```

Note: As we can see that this is a long vector produced by the sapply function which is not readable, to avoide this kind of mistake we need to use a dataframe by using data.frame() on Datasets.

```r
veg_means = sapply(data.frame(Datasets), FUN=mean)
```

```r
# Now checking the mean of the datas
veg_means
```

Note that we have lost the name of the columns, to avoid this kind of mistake we need to add the column name by using the colnames() method on the dataframe.

```r
# adding  column names 
# we need to add the column on the output of the cbind method before makig it a dataframe object.
Datasets = cbind(veg$R, veg$ROCK, veg$LITTER, veg$ML, veg$BARESOIL)
colnames(Datasets) = c("R", "ROCK", "LITTER", "ML", "BARESOIL")
```

```r
# 
veg_means = sapply(data.frame(Datasets), FUN=mean)
veg_means
```
